```python
def get_album_images(genre):
  off_playl=0
  playlists = sp.category_playlists(genre, limit=50, offset=0)['playlists']
  while playlists['next'] is not "None":
    off_playl += 50
    playlists = sp.category_playlists(genre, limit=50, offset=off_playl)['playlists'] # get more playlists
  trackSet = set()
  #off=0
  for playlist in playlists['items']: 
    tracks = sp.playlist_tracks(playlist['id'], limit=100, offset=0) # get more tracks from each playlist
    #off += 100
    for track in tracks['items']:
      albumImages = track['track']['album']['images']
      firstImageUrl = albumImages[0]['url']
      trackSet.add(firstImageUrl)
      
  return trackSet
```
